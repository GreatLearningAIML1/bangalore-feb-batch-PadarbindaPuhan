Question # 4 Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference. 

```python
# Method - 1
def Goal_diff_count(GoalsFor, GoalsAgainst):
    return (GoalsFor - GoalsAgainst)

print('Team with max Goal Diff:',LaLiga.loc[Goal_diff_count((LaLiga['GoalsFor']).astype(int),LaLiga['GoalsAgainst'].astype(int)).idxmax()]['Team'])
print('Team with max Goal Diff:',LaLiga.loc[Goal_diff_count((LaLiga['GoalsFor']).astype(int),LaLiga['GoalsAgainst'].astype(int)).idxmin()]['Team'])
```

```python
# Method - 2
# This is how we make life more complex

df1 = pd.DataFrame(columns=['Team','goal_diff'])

def Goal_diff_count(Team,GoalsFor, GoalsAgainst,row_index):
    goal_diff = int(GoalsFor) - int(GoalsAgainst)
    df1.loc[row_index] = [Team,goal_diff]

for row_index,row in LaLiga.iterrows():
    Goal_diff_count(row['Team'],row['GoalsFor'],row['GoalsAgainst'],row_index)
    
df1['goal_diff'] = pd.to_numeric(df1['goal_diff'])

print ('Team with max goal diff:',df1.loc[df1['goal_diff'].idxmax()]['Team'])
print ('Team with max goal diff:',df1.loc[df1['goal_diff'].idxmin()]['Team'])
```

Question # 5 Create a new column with name “Winning Percent” and append it to the data set (5 points)
Percentage of Winning = (GamesWon / GamesPlayed)*100

If there are any numerical error, replace it with 0%

Print the top 5 teams which has the highest Winning percentage

```python
# Method 1 - using sort_values
LaLiga['Winning Percent'] = (LaLiga['GamesWon'].astype(int) / LaLiga['GamesPlayed'].astype(int))*100
LaLiga.fillna(0,inplace= True)
LaLiga.sort_values('Winning Percent',axis = 0 , ascending = False)[['Team','Winning Percent']].head(5)
```

```python
#Method 2 - using nlargest
LaLiga['Winning Percent'] = (LaLiga['GamesWon'].astype(int) / LaLiga['GamesPlayed'].astype(int))*100
LaLiga.fillna(0,inplace= True)
LaLiga.nlargest(5,['Winning Percent'])[['Team','Winning Percent']]
```
