#### Do train and test split with stratify sampling on Outcome variable to maintain the distribution of dependent variable(1 Marks)

```python
from sklearn.model_selection import train_test_split
X = diab_df.drop("Outcome", axis=1)
y = diab_df["Outcome"]
test_size = 0.30 # taking 70:30 training and test set
seed = 7  # Random numbmer seeding for reapeatability of the code
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed,stratify=y)
```

### Train Support Vector Machine Model(1 Marks)

```python
# use from sklearn.svm import SVC
#Cheking for defult values below. 
```

```python
from sklearn.svm import SVC
model =SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
# Default kernel is choosen as rbf and C = 1
#Lets check the score with both rbf and linear kernel with C = 1
```

```python
ker = ['rbf','linear']
for k in ker:
    model = SVC(kernel = k)
    model.fit(X_train, y_train)
    print(k,'-','train -',model.score(X_train, y_train))
    print(k,'-','test -',model.score(X_test, y_test))
```

### Scale the data points using MinMaxScaler ( 1 Marks)

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_X=scaler.fit_transform(X)
scaled_X
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, test_size=test_size, random_state=seed)
```

#we can try increasing either C or gamma to fit a more complex model.

### Try improving the model accuracy using C=1000 (1 Marks)

```python
ker = ['rbf','linear']
for k in ker:
    model = SVC(C=1000,kernel = k)
    model.fit(X_train, y_train)
    print(k,'-',model.score(X_train, y_train))
    print(k,'-',model.score(X_test, y_test))
```

```python
ker = ['rbf','linear']
gamma = [10,100,500,1000,2000,5000]
res = pd.DataFrame()
for k in ker:
    for g in gamma:
        model = SVC(C=g,kernel = k)
        model.fit(X_train, y_train)
        res = res.append({'kernel':k,'C':g,'accuracy':model.score(X_test, y_test)},ignore_index=True)        
```

```python
res
```

```python
#After increasing C, result for both the kernels has improved. 
#For rbf accuracy score is 0.78 for C = 2000
#For linear kernel accuracy score is 0.76 for C = 500. There is no increase in score with further increase in C. 
```
